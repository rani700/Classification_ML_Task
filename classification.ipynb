```python
import pandas as pd
```

```python
df = pd.read_csv("/home/shweta/train.csv")
```

```python
df
```

```python
x_train = df.iloc[:,:-1]
```

```python
y_train = df.target
```

```python
df1 = pd.read_csv("/home/shweta/test.csv")
```

```python
x_test = df1
```

```python
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(x_train)
```

```python
X_test = StandardScaler().fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression(max_iter = 1000)
```

```python
logReg.fit(X_train,y_train)
```

```python
y_log = logReg.predict(X_test)
```

```python
logReg.score(X_train, y_train)
```

```python
df1['target_LogReg'] = y_log
```

```python
df1
```

```python
from sklearn.naive_bayes import GaussianNB
gau = GaussianNB()
gau.fit(X_train, y_train)
```

```python
y_gau = gau.predict(X_test)
```

```python
gau.score(X_train, y_train)
```

```python
df1['target_gaussian'] = y_gau
```

```python
df1
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train,y_train)
```

```python
knn.score(X_train,y_train)
```

```python
y_knn = knn.predict(X_test)
```

```python
df1["target_knn"] = y_knn
```

```python
df1
```

```python
from sklearn import svm
sv = svm.SVC()
sv.fit(X_train, y_train)
```

```python
y = sv.predict(X_test)
```

```python
sv.score(X_train, y_train)
```

```python
df1['target_svm'] = y
```

```python
df1
```

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_dt = dt.predict(X_test)
```

```python
dt.score(X_train, y_train)
```

```python
df1['target_DecisionTree'] = y_dt
```

```python
df1
```
